```python
#distplot 

tips
```

```python
sns.displot(tips[tips['sex']=='Male']['total_bill'])
plt.xlabel('total_bills_by_male')
plt.savefig('d-tips')
```

```python
sns.displot(tips[tips['sex']=='Female']['total_bill'])
plt.xlabel('total_bills_by_female')
plt.savefig('d-tip2')
```

```python

sns.displot(tips[tips['smoker']=='Yes']['total_bill'])
plt.xlabel('total_bills_by_smoker')
plt.savefig('d_tips')
plt.show()
```

```python
sns.distplot(tips[tips['smoker']=='Yes']['total_bill'],hist=False,label='Smoker')
sns.distplot(tips[tips['smoker']=='No']['total_bill'],hist=False,label='Non-Smoker')
plt.legend()
plt.xlabel('total_bills_by_smoker_and_nonsmoker')
plt.savefig('smoker_VS_nonSmoker')
plt.show()
```

```python
sns.distplot(tips[tips['smoker']=='Yes']['tip'],hist=False,label='Smoker')
sns.distplot(tips[tips['smoker']=='No']['tip'],hist=False,label='Non-Smoker')

plt.legend()
plt.xlabel('tips_by_smoker_and_nonsmoker')
plt.savefig('smoker_VS_nonSmoker2')
plt.show()
```

```python
sns.distplot(tips[tips['sex']=='Male']['total_bill'],hist=False,label='Male')
sns.distplot(tips[tips['sex']=='Female']['total_bill'],hist=False,label='Female')
plt.legend()
plt.xlabel('total_bills_by_Male_and_Female')
plt.savefig('Male_VS_Female3')
plt.show()
```

```python
sns.distplot(tips[tips['sex']=='Male']['tip'],hist=False,label='Male')
sns.distplot(tips[tips['sex']=='Female']['tip'],hist=False,label='Female')
plt.legend()
plt.xlabel('Tips_by_Male_and_Female')
plt.savefig('Male_VS_Female2')
plt.show()
```
